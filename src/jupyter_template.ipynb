```python
from mirror.models.mirror_llama_model import MirrorLlamaModel
from mirror.models.configuration_llama import LlamaConfig
from mirror.datasets.wikitext_dataset import WikitextDataset
from mirror.trainer import Trainer
from lightning.fabric.strategies.single_device import SingleDeviceStrategy
```

```python
model = MirrorLlamaModel(initialization=LlamaConfig(
    vocab_size=128256,
    hidden_size=128,
    intermediate_size=340,
    num_hidden_layers=6,
    num_attention_heads=4,
    num_key_value_heads=None,
))

# model = MirrorLlamaModel(initialization='3.2-1B-Instruct')
# model = MirrorGPTModel(weights='pretrained')

dataset = WikitextDataset(split="train", head=10)

from mirror.config import get_config
get_config()['device'] = 'cuda'

trainer = Trainer(strategy=SingleDeviceStrategy(device="cuda:0"))
```

```python
trainer.fit(model, dataset, epochs=1, batch_size=1)
```
